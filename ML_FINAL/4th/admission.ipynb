```python
x = df[['GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'LOR ', 'CGPA',
       """
       This code snippet selects specific columns from a DataFrame `df` to create feature and target variables for a machine learning model.

       Variables:
              x (DataFrame): A DataFrame containing the following columns as features:
                     - 'GRE Score': Graduate Record Examination score.
                     - 'TOEFL Score': Test of English as a Foreign Language score.
                     - 'University Rating': Rating of the university.
                     - 'SOP': Statement of Purpose strength.
                     - 'LOR ': Letter of Recommendation strength.
                     - 'CGPA': Cumulative Grade Point Average.
                     - 'Research': Research experience (1 if yes, 0 if no).

              y (Series): A Series containing the target variable:
                     - 'Chance of Admit ': The probability of admission.
       """
       'Research']]

y = df['Chance of Admit ']
```

```python
from sklearn.model_selection import train_test_split
"""
This script is part of a Jupyter Notebook located at:

It imports the `train_test_split` function from the `sklearn.model_selection` module.
The `train_test_split` function is used to split arrays or matrices into random train and test subsets.
"""
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=1)
"""
Splits the dataset into training and testing sets.

Parameters:
x : array-like or DataFrame
    The input features for the dataset.
y : array-like or Series
    The target variable for the dataset.
test_size : float, optional (default=0.25)
    The proportion of the dataset to include in the test split.
random_state : int, optional (default=1)
    Controls the shuffling applied to the data before applying the split.
    Pass an int for reproducible output across multiple function calls.

Returns:
x_train : array-like or DataFrame
    The training input features.
x_test : array-like or DataFrame
    The testing input features.
y_train : array-like or Series
    The training target variable.
y_test : array-like or Series
    The testing target variable.
"""
```

```python
print(f"Size of splitted data")
"""
This code snippet prints the size of the split data for a machine learning model.

The following variables are expected to be defined before this code is executed:
- x_train: The training data features.
- y_train: The training data labels.
- x_test: The testing data features.
- y_test: The testing data labels.

The output will display the shapes of these datasets, which helps in verifying that the data has been split correctly.

Output:
- Size of splitted data
- x_train: Shape of the training data features.
- y_train: Shape of the training data labels.
- x_test: Shape of the testing data features.
- y_test: Shape of the testing data labels.
"""
print(f"x_train {x_train.shape}")
print(f"y_train {y_train.shape}")
print(f"y_train {x_test.shape}")
print(f"y_test {y_test.shape}")
```

```python

"""
This script imports three machine learning models from the scikit-learn library:

1. DecisionTreeRegressor: A decision tree regressor is a decision tree that performs regression tasks. 
    It predicts the value of a target variable by learning simple decision rules inferred from the data features.

2. RandomForestRegressor: A random forest regressor is an ensemble learning method that operates by constructing 
    multiple decision trees during training and outputting the mean prediction of the individual trees. 
    It improves predictive accuracy and controls over-fitting.

3. LogisticRegression: Despite its name, logistic regression is a linear model for binary classification. 
    It estimates the probability that a given input point belongs to a certain class.
"""
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LogisticRegression
```

```python

"""
This code initializes three different machine learning models with a specified random state for reproducibility.

Models:
- DecisionTreeRegressor: A decision tree regressor model used for predicting continuous values.
- RandomForestRegressor: A random forest regressor model, which is an ensemble of decision trees, used for predicting continuous values.
- LogisticRegression: A logistic regression model used for binary classification tasks.

Parameters:
- random_state: An integer value used to seed the random number generator for reproducibility of results.
- solver: The algorithm to use in the optimization problem for LogisticRegression. 'lbfgs' is an optimization algorithm that handles multinomial loss.
- max_iter: The maximum number of iterations taken for the solvers to converge in LogisticRegression.
"""
model_dt = DecisionTreeRegressor(random_state=1)
model_rf = RandomForestRegressor(random_state=1)
model_lr = LogisticRegression(random_state=1,solver='lbfgs',max_iter=1000)
```

```python
model_dt.fit(x_train,y_train)
"""
Fits the decision tree model using the training data.

This method trains the decision tree classifier on the provided training data.
The model learns the relationship between the features (x_train) and the target variable (y_train).

Parameters:
x_train (array-like or DataFrame): The training input samples. Shape (n_samples, n_features).
y_train (array-like or Series): The target values (class labels) corresponding to x_train. Shape (n_samples,).

Returns:
None: This method does not return anything. It updates the model in place.
"""
```

```python
model_rf.fit(x_train,y_train)
"""
Fits the Random Forest model to the training data.

This method trains the Random Forest model using the provided training features and labels.

Parameters:
x_train : array-like or DataFrame of shape (n_samples, n_features)
    The training input samples. Each row represents a sample, and each column represents a feature.
    
y_train : array-like of shape (n_samples,)
    The target values (class labels in classification, real numbers in regression).

Returns:
None
"""
```

```python
model_lr.fit(x_train,y_train)
"""
Fits the logistic regression model using the training data.

This method trains the logistic regression model (`model_lr`) using the provided
training features (`x_train`) and training labels (`y_train`).

Parameters:
    x_train (array-like or DataFrame): The training input samples. This should be
        a 2D array or DataFrame where each row represents a sample and each column
        represents a feature.
    y_train (array-like or Series): The target values (class labels) corresponding
        to the training input samples. This should be a 1D array or Series.

Returns:
    None: This method does not return any value. It updates the state of `model_lr`
    by fitting it to the training data.
"""
```

```python
y_pred_dt = model_dt.predict(x_test) #int
"""
Predict the target values using different models.

This code snippet demonstrates how to use three different machine learning models
to predict the target values for a given test dataset `x_test`.

Variables:
    y_pred_dt (numpy.ndarray): Predicted target values using the Decision Tree model. The predictions are integers.
    y_pred_rf (numpy.ndarray): Predicted target values using the Random Forest model. The predictions are floats.
    y_pred_lr (numpy.ndarray): Predicted target values using the Logistic Regression model. The type of predictions is not specified.

Models:
    model_dt: Trained Decision Tree model.
    model_rf: Trained Random Forest model.
    model_lr: Trained Logistic Regression model.

Usage:
    y_pred_dt = model_dt.predict(x_test)
    y_pred_rf = model_rf.predict(x_test)
    y_pred_lr = model_lr.predict(x_test)
"""
y_pred_rf = model_rf.predict(x_test) #float
y_pred_lr = model_lr.predict(x_test) #
```

```python
y_pred_rf = [1 if each > 0.75 else 0 for each in y_pred_rf]
"""
Convert predicted probabilities to binary classification.

This line of code iterates over the list `y_pred_rf`, which contains predicted probabilities 
from a random forest classifier. For each probability, it assigns a binary value:
- 1 if the probability is greater than 0.75 (indicating a positive class prediction)
- 0 if the probability is 0.75 or less (indicating a negative class prediction)

Returns:
    list: A list of binary values (0 or 1) based on the threshold of 0.75.
"""
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, accuracy_score
"""
This script imports necessary modules from the scikit-learn library for evaluating machine learning models.

Modules imported:
- ConfusionMatrixDisplay: A utility for visualizing the confusion matrix.
- accuracy_score: A function to compute the accuracy of the model.
- classification_report: A function to generate a detailed classification report including precision, recall, F1-score, and support.

Usage:
These imports are typically used in the context of evaluating the performance of a classification model. 
For example, after training a model, you can use these tools to visualize the confusion matrix, 
calculate the accuracy, and generate a comprehensive classification report.
"""
from sklearn.metrics import classification_report
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred_dt)
"""
Displays the confusion matrix for the decision tree model predictions, 
sets the title of the plot, and prints the accuracy and classification report.

Steps:
1. Displays the confusion matrix using `ConfusionMatrixDisplay.from_predictions`.
2. Sets the title of the plot to 'Decision Tree'.
3. Shows the plot.
4. Prints the accuracy of the model using `accuracy_score`.
5. Prints the classification report using `classification_report`.

Parameters:
- y_test: array-like of shape (n_samples,)
    True labels for the test data.
- y_pred_dt: array-like of shape (n_samples,)
    Predicted labels by the decision tree model.
"""
plt.title('Decision Tree')
plt.show()
print(f" Accuracy is {accuracy_score(y_test,y_pred_dt)}")
print(classification_report(y_test,y_pred_dt))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred_lr)
"""
Displays the confusion matrix for logistic regression predictions, 
plots the confusion matrix with a title, and prints the accuracy 
and classification report.

Functions:
    ConfusionMatrixDisplay.from_predictions(y_test, y_pred_lr): 
        Generates the confusion matrix from the true labels and predicted labels.
    plt.title('Logistic Regression'): 
        Sets the title of the plot to 'Logistic Regression'.
    plt.show(): 
        Displays the plot.
    print(f" Accuracy is {accuracy_score(y_test, y_pred_lr)}"): 
        Prints the accuracy score of the logistic regression model.
    print(classification_report(y_test, y_pred_lr)): 
        Prints the classification report, which includes precision, recall, 
        f1-score, and support for each class.
        
Parameters:
    y_test (array-like): True labels of the test dataset.
    y_pred_lr (array-like): Predicted labels from the logistic regression model.
"""
plt.title('Logistic Regression')
plt.show()
print(f" Accuracy is {accuracy_score(y_test,y_pred_lr)}")
print(classification_report(y_test,y_pred_lr))
```

```python
#Random Forest
"""
This code snippet visualizes the performance of a Random Forest classifier using a confusion matrix and prints the accuracy and classification report.

Steps:
1. `ConfusionMatrixDisplay.from_predictions(y_test, y_pred_rf, xticks_rotation='vertical')`: 
    - This line generates a confusion matrix from the true labels (`y_test`) and the predicted labels (`y_pred_rf`) of the Random Forest model.
    - The `xticks_rotation='vertical'` argument rotates the x-axis labels vertically for better readability.

2. `plt.title('Random Forest')`: 
    - Sets the title of the plot to 'Random Forest'.

3. `plt.show()`: 
    - Displays the confusion matrix plot.

4. `print(f" Accuracy is {accuracy_score(y_test, y_pred_rf)}")`: 
    - Calculates and prints the accuracy of the Random Forest model using the `accuracy_score` function.

5. `print(classification_report(y_test, y_pred_rf))`: 
    - Prints the classification report, which includes precision, recall, f1-score, and support for each class.
"""
ConfusionMatrixDisplay.from_predictions(y_test,y_pred_rf,xticks_rotation='vertical')
plt.title('Random Forest')
plt.show()
print(f" Accuracy is {accuracy_score(y_test,y_pred_rf)}")
print(classification_report(y_test,y_pred_rf))
```
