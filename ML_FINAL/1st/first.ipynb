```python
#removing duplicates  #8
"""
This script removes duplicate rows from a dataset.

Functions:
    - drop_duplicates: Removes duplicate rows from the dataset.

Variables:
    - dataset_cleaned: A DataFrame containing the dataset with duplicates removed.
"""
dataset_cleaned=dataset.drop_duplicates()
```

```python
#check for missing values  #10
"""
This script checks for missing values in the cleaned dataset and prints the count of missing values for each column.

Steps:
1. Calculate the number of missing values in each column of the dataset.
2. Print the count of missing values for each column.

Variables:
- dataset_cleaned: The cleaned dataset (assumed to be a pandas DataFrame).
- missing_values: A pandas Series containing the count of missing values for each column in the dataset.
"""
missing_values = dataset_cleaned.isnull().sum()
print(missing_values)
```

```python
#save the cleaned data to the new csv file #11
"""
Saves the cleaned dataset to a new CSV file.

This function takes the cleaned dataset and saves it to a CSV file named 'heart_cleaned.csv' 
in the current working directory. The index of the DataFrame is not included in the saved file.

Parameters:
None

Returns:
None

Example:
    dataset_cleaned.to_csv('heart_cleaned.csv', index=False)
"""
dataset_cleaned.to_csv('heart_cleaned.csv',index=False)
```

```python
#handle missing data by filling with mean/median value of respective columns
"""
This script handles missing data in a dataset by filling missing values with the median value of the respective columns.

For each column in the dataset:
- If the column contains any missing values (NaNs):
    - Calculate the median value of the column.
    - Replace all missing values in the column with the calculated median value.

Parameters:
- dataset (pandas.DataFrame): The dataset containing columns with potential missing values.

Returns:
- None: The dataset is modified in place.
"""
#12
for column in dataset.columns:
    if dataset[column].isnull().sum() > 0: # 1>0 true  and 0>0 false
        median_value = dataset[column].median()
        dataset[column].fillna(median_value, inplace=True)
```

```python
#verify no missing values remain #13
"""
This script verifies that no missing values remain in the dataset after handling them.

Steps:
1. Calculate the number of missing values in each column of the dataset after handling.
2. Print the number of missing values in each column.

Variables:
- missing_values_after: A pandas Series containing the count of missing values for each column in the dataset.

Output:
- Prints the number of missing values in each column after handling.
"""

missing_values_after=dataset.isnull().sum()
print("missing values in each column after handling:")
print(missing_values_after)
```

```python
#select the columns in the dataset that have numerical data types  #14
"""
Select numerical columns from the dataset.

This function selects the columns in the dataset that have numerical data types 
(i.e., integer and float) and stores their names in the variable `numerical_features`.

Attributes:
    dataset (DataFrame): The input dataset from which numerical columns are to be selected.

Returns:
    Index: An Index object containing the names of the columns with numerical data types.
"""
numerical_features=dataset.select_dtypes(include=['int64','float64']).columns
```

```python
# Initialize the StandardScaler #16
# The StandardScaler standardizes features by removing the mean and scaling to unit variance.
# This is useful for ensuring that each feature contributes equally to the model's performance.
from sklearn.preprocessing import StandardScaler

# Create an instance of the StandardScaler
# This instance will be used to fit and transform the dataset's numerical features.
scaler = StandardScaler()
```

```python
#16 Fit the scaler to your data: This means calculating the average value (mean) and how spread out the values are (standard deviation) for each column in your data.
"""
Fits the scaler to the dataset.

This function calculates the mean and standard deviation for each column in the dataset.
These statistics are used to scale the data, typically for normalization or standardization purposes.

Parameters:
dataset (array-like): The data to fit the scaler to. This should be an array-like structure 
                      (e.g., list, NumPy array, pandas DataFrame) where each column represents 
                      a feature and each row represents an observation.

Returns:
None: This function does not return anything. It updates the internal state of the scaler 
      with the calculated mean and standard deviation for each column.
"""
scaler.fit(dataset)
```

```python
#fit and transform the numerical features
"""
Fits the scaler to the numerical features of the dataset and transforms these features.

This process standardizes the numerical features by removing the mean and scaling to unit variance.

Parameters:
    dataset (pd.DataFrame): The dataset containing the features to be scaled.
    numerical_features (list): A list of column names in the dataset that are numerical features.
    scaler (object): An instance of a scaler (e.g., StandardScaler) that has fit and transform methods.

Returns:
    pd.DataFrame: The dataset with the numerical features scaled.
"""
dataset[numerical_features]=scaler.fit_transform(dataset[numerical_features])
```

```python
# Display the transformed data #17
# Scaling means adjusting the numbers in your data so they're on a similar scale.
# This is useful for ensuring that each feature contributes equally to the model's performance.

# Print a message indicating that the data after scaling will be displayed
print("Data after scaling:")

# Use the head() method to print the first few rows of the scaled dataset.
# The head() method by default prints the first 5 rows of the DataFrame.
print(dataset.head())
```

```python
#18 D. Randomly divide dataset into training (75%) and testing (25%) sets
# Import the train_test_split function from sklearn.model_selection
from sklearn.model_selection import train_test_split

# Use train_test_split to split the dataset into training and testing sets
# Parameters:
# - dataset: The dataset to be split.
# - test_size=0.25: Specifies that 25% of the data should be used for testing.
# - random_state=42: Ensures reproducibility of the split by setting a seed for the random number generator.
train_data, test_data = train_test_split(dataset, test_size=0.25, random_state=42)

# The resulting train_data and test_data are now available for further analysis or model training/testing.
```
