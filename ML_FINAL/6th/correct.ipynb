```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans, AgglomerativeClustering

#This cell imports the necessary libraries for the clustering analysis. 
#The numpy and pandas libraries are used for data manipulation and processing.
#The matplotlib and seaborn libraries are used for data visualization.
#The KMeans and AgglomerativeClustering classes from the sklearn.cluster module are used for the clustering algorithms.


```

```python
# Loading dataset
df = pd.read_csv(r"E:\TY\ML\ML LAB\MLCODES\6th\Mall_Customers.csv")
df.head()
"""
This cell loads the mall customer dataset from the provided path and displays the first few rows of the data.
The dataset contains information about the customers, including their gender, age, annual income, and spending score.
"""
```

```python
# Selecting features for clustering
x = df.iloc[:, 3:]
"""
This cell selects the features that will be used for the clustering analysis, which are the annual income and spending score columns.
The iloc function is used to select the columns from index 3 onwards (annual income and spending score).
"""
```

```python
plt.title('Unclustered Data')
"""
This code snippet sets the title of a plot to 'Unclustered Data' and creates a scatter plot using seaborn.

- The x-axis represents 'Annual Income (k$)' from the DataFrame `x`.
- The y-axis represents 'Spending Score (1-100)' from the DataFrame `x`.

Libraries used:
- matplotlib.pyplot as plt
- seaborn as sns
"""
sns.scatterplot(x=x['Annual Income (k$)'],y=x['Spending Score (1-100)'])
```

```python
km = KMeans(n_clusters=5,random_state=1)
labels = km.fit_predict(x)
km.labels_
cent = km.cluster_centers_
plt.title('Clustered Data')

sns.scatterplot(x=x['Annual Income (k$)'], y=x['Spending Score (1-100)'], hue=labels, palette='viridis')
sns.scatterplot(x=cent[:,0], y=cent[:,1], s=200, color='red', label='Centroids')
"""
This cell applies the KMeans clustering algorithm to the dataset.
The n_clusters parameter is set to 5, indicating that the algorithm should partition the data into 5 clusters.
The fit_predict method is used to fit the KMeans model and predict the cluster labels for each data point.

This cell creates a scatter plot of the data points, colored by their assigned cluster labels from the KMeans algorithm.
The cluster centroids are also plotted as red dots on the scatter plot.
The viridis color palette is used to assign colors to the different clusters.
"""
```

```python
# Agglomerative Clustering
agl = AgglomerativeClustering(n_clusters=5)
alabels = agl.fit_predict(x)
alabels
"""
This cell applies the Agglomerative Clustering algorithm to the dataset.
The n_clusters parameter is set to 5, indicating that the algorithm should partition the data into 5 clusters.
The fit_predict method is used to fit the Agglomerative Clustering model and predict the cluster labels for each data point.
The predicted cluster labels are stored in the labels_agg variable.
"""
```

```python

"""
This code generates a figure with two subplots comparing the results of Agglomerative Clustering and K-Means Clustering on a dataset.

Subplots:
1. Agglomerative Clustering:
    - Plots a scatter plot of 'Annual Income (k$)' vs 'Spending Score (1-100)'.
    - Colors the points based on the agglomerative clustering labels (`alabels`).

2. K-Means Clustering:
    - Plots a scatter plot of 'Annual Income (k$)' vs 'Spending Score (1-100)'.
    - Colors the points based on the K-Means clustering labels (`labels`).
    - Overlays the centroids of the clusters in red with a larger size (`s=200`).

Parameters:
- x: DataFrame containing the data with columns 'Annual Income (k$)' and 'Spending Score (1-100)'.
- alabels: Array-like, cluster labels from Agglomerative Clustering.
- labels: Array-like, cluster labels from K-Means Clustering.
- cent: Array-like, coordinates of the centroids from K-Means Clustering.
"""
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title('Agglomerative')
sns.scatterplot(x=x['Annual Income (k$)'], y=x['Spending Score (1-100)'], hue=labels)

plt.subplot(1,2,2)
plt.title('KMEANS')
sns.scatterplot(x=x['Annual Income (k$)'], y=x['Spending Score (1-100)'], hue=labels)
sns.scatterplot(x=cent[:,0], y=cent[:,1], s=200, color='red', label='Centroids')
```
